```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_montecarlo
```

```python
file_data = 'ANN_training_Newjig_0213_all_LM_result.csv'
df = pd.read_csv(file_data, encoding="gb18030")
df.head()
```

```python
mc = df['total_pre_y'].montecarlo(sims=10, bust=-0.1, goal=1)
```

```python
mc.plot(title="SPY Returns Monte Carlo Simulations")  # optional: , figsize=(x, y)
df['total_pre_y'].plot()
```
